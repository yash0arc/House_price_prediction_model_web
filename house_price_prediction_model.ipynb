```python
def total_sqft_cleaning(x):
    try:
        parts = str(x).split(" ")
        if len(parts) == 3:
            y, _, z = parts
            return (float(y) + float(z)) / 2
        elif len(parts) == 1:
            return float(parts[0])
        else:
            return np.nan
    except:
        return np.nan
```

```python
df2.total_sqft=df2.total_sqft.apply(lambda x :total_sqft_cleaning(x))
df2.total_sqft
```

```python
x=df3.drop(['area_type','location','price'],axis=1)
y=df3.price
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
models = {
    LinearRegression: {
        'model': LinearRegression(),
        'params': {}
    },
    Ridge: {
        'model': Ridge(),
        'params': {
            'alpha': [0.1, 1.0, 10.0]
        }
    },
    Lasso: {
        'model': Lasso(),
        'params': {
            'alpha': [0.01, 0.1, 1.0]
        }
    }
}
```

```python

for name,m in models.items():
    model=m['model']
    params=m['params']
    grid=GridSearchCV(model,params,cv=5)
    grid.fit(x,y)
    print(name)
    print(grid.best_score_*100)
    print(grid.best_params_)
```

```python
len(df3)
```

```python
#bathrooms should not be more than no of bedrooms
df3=df3[df3['bath'] <= df3['size'] + 1]
```

```python
df3['sq_per_room']=df3['total_sqft']/df3['size']
df3
```

```python
df4=df3[(df3['sq_per_room']>=100) & (df3['sq_per_room'] <= 650)]
len(df4)
```

```python
df4.describe()
```

```python

for col in ['size','bath','price']:
    Q1 = df4[col].quantile(0.25)
    Q3 = df4[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df5 = df4[(df4['sq_per_room'] >= lower_bound) & (df4['sq_per_room'] <= upper_bound)]
df5.describe()
```

```python
xtrain=df5.drop(['area_type','location','price'],axis=1)
ytrain=df5.price
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
models = {
    LinearRegression: {
        'model': LinearRegression(),
        'params': {}
    },
    Ridge: {
        'model': Ridge(),
        'params': {
            'alpha': [0.1, 1.0, 10.0]
        }
    },
    Lasso: {
        'model': Lasso(),
        'params': {
            'alpha': [0.01, 0.1, 1.0]
        }
    }
}

for name,m in models.items():
    model=m['model']
    params=m['params']
    grid=GridSearchCV(model,params,cv=5)
    grid.fit(xtrain,ytrain)
    print(name)
    print(grid.best_score_*100)
    print(grid.best_params_)
```

```python
model=Ridge(alpha=1)
model.fit(xtrain,ytrain)
model.score(xtrain,ytrain)*100
```

```python
def prediction(location, total_sqft, areatype, bedrooms, bathrooms):
    x_input = np.zeros(len(xtrain.columns))  
    x_input[0] = bedrooms
    x_input[1] = total_sqft
    x_input[2] = bathrooms

    loc_col = location
    if loc_col in columns:
        loc_index = columns.get_loc(loc_col)
        x_input[loc_index] = 1

    area_col = areatype
    if area_col in columns:
        area_index = columns.get_loc(area_col)
        x_input[area_index] = 1

    x_df = pd.DataFrame([x_input], columns=columns)
    prediction_value = model.predict(x_df)
    print("Predicted Price:", prediction_value[0]*100000)
prediction('Vittasandra',1500,'Super built-up  Area',3,3)
```
